```python
import pyautogui
import time

# Move the mouse to coordinates (x, y) over the duration of seconds
def move_mouse_to(x, y, duration=1):
    pyautogui.moveTo(x, y, duration=duration)

# Move the mouse relative to the current position
def move_mouse_relative(dx, dy, duration=1):
    pyautogui.move(dx, dy, duration=duration)

# Example usage
if __name__ == "__main__":
    # Move the mouse to (100, 100) over 2 seconds
    move_mouse_to(100, 0, duration=2)

    # Move the mouse 50 pixels right and 50 pixels down over 1 second
    move_mouse_relative(50, 0, duration=1)
```
